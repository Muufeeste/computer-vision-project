# Resizing images

```python
resized = cv2.resize(image, (200, 200))

cv2_imshow(resized)
```

```python
r = 300.0 / w
dim = (300, int(h * r))
resized = cv2.resize(image, dim)
cv2_imshow(resized)
```

```python
r = 150.0 / w
dim = (150, int(h * r))
resized = cv2.resize(image, dim)
cv2_imshow(resized)
```

```python
import imutils
```

```python
resized = imutils.resize(image, width=150)

cv2_imshow(resized)
```

# Rotation images

```python
center = (w // 2, h // 2)
rotation_matrix = cv2.getRotationMatrix2D(center, -45, 1.0)
rotated = cv2.warpAffine(image, rotation_matrix, (w, h))

cv2_imshow(rotated)
```

```python
rotated = imutils.rotate(image, -45)

cv2_imshow(rotated)
```

```python
rotated = imutils.rotate_bound(image, 45)

cv2_imshow(rotated)
```

# Drawing and writing on the picture

```python
output = image.copy()
cv2.rectangle(output, (310, 90), (350, 150), (0, 0, 255), 3)

cv2_imshow(output)
```

```python
cv2.circle(output, (300, 150), 10, (255, 0, 0), -1)

cv2_imshow(output)
```

```python
output = image.copy()
cv2.line(output, (60, 20), (400, 200), (0, 0, 255), 5)

cv2_imshow(output)
```

```python
output = image.copy()
cv2.putText(output, "Hello World", (10, 25), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)

cv2_imshow(output)
```

# Convert image to grayscale and detect edges

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

cv2_imshow(gray)
```

```python
edged = cv2.Canny(image, 30, 150)

cv2_imshow(edged)
```
